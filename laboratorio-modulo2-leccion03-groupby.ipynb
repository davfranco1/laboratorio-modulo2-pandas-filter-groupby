```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None) 

df = pd.read_csv("netflix_merge.csv")

df.head(1)
```

#### Ejercicio 1: ¿Cuántos títulos hay por cada tipo de contenido ('Movie' o 'TV Show')?

```python
df.groupby("type").count()["title"].reset_index()
```

#### Ejercicio 2: ¿Cuál es el promedio de puntuación de las películas agrupadas por clasificación de audiencia ('rating')? Interpreta los resultados, ¿hay algo que te llame la atención? Explica por qué. 

```python
avg = df.groupby("rating")["IMDB Score"].mean().reset_index()

condicion_avg_notnull = avg.notnull()
avg[condicion_avg_notnull]
```

#### Ejercicio 3: ¿Cuántos títulos hay por cada país? Ordena los resultados de mayor a menor

```python
df_paises = df.groupby("country").count()["title"].sort_values(ascending=False).reset_index()
df_paises
```

#### Ejercicio 4: ¿Cuál es la calificación promedio en IMDb de los títulos agrupados por tipo ('Movie' o 'TV Show')?

```python
avg = df.groupby("type")["IMDB Score"].mean().reset_index()

condicion_tipo = avg.notnull()
avg[condicion_tipo]
```

#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python
df["date_added"]=pd.to_datetime(df["date_added"].str.strip(), format="%B %d, %Y")
```

```python
df["Año"] = df["date_added"].dt.year
df_año = df.groupby("Año")[["title"]].count()
df_año
```

#### Ejercicio 6: ¿Cuál es el país con más títulos lanzados en el año 2020? Para resolver este ejercicio después del gropuby os puede resultar interesante el método [`idxmax`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html)

```python
df_año_2020 = df[df["release_year"] == 2020]
lista_paises = df_año_2020.groupby("country")[["title"]].count()
lista_paises.sort_values(by = "country", ascending=False)

```

#### Ejercicio 7:¿Cuál es el promedio de puntuación en IMDb de los títulos agrupados por país?

```python
promedio_imdb = df.groupby("country")["IMDB Score"].mean()
condicion_imdb = promedio_imdb.notnull()

promedio_imdb[condicion_imdb].sort_values(ascending=False).reset_index()
```

#### Ejercicio 8: ¿Cuántos títulos fueron lanzados en cada año agrupados por tipo ('Movie' o 'TV Show')?

```python
df_tipo = df.groupby(["type", "release_year"])[["title"]].count()
df_tipo
```

#### Ejercicio 9: ¿Cuál es el país con la mayor cantidad de títulos de clasificación 'TV-MA'?

```python
condicion_tvma = df["rating"]=="TV-MA"
df_condicion_tvma = df[condicion_tvma]
df_condicion_tvma.groupby("country")["title"].count().nlargest(1).reset_index()
```

#### Ejercicio 10. ¿Cuál es el promedio de duración de las películas agrupadas por país?

```python
df_duracion_pelis = df[df["type"] == "Movie"]
df_duracion_pelis.groupby("country")[["Runtime"]].mean().sort_values(by = "country", ascending=True)
```

#### Ejercicio 11: ¿Cuál es el número total de títulos por país y clasificación de audiencia ('rating')?

```python
df_pais_rating = df.groupby(["country","rating"])[["title"]].count().sort_values(by = "title", ascending=False)
df_pais_rating.head()
```

#### Ejercicio 12: ¿Cuántos títulos fueron lanzados antes del año 2000 agrupados por país?

```python
df_antes_2000 = df[df["release_year"] < 2000]
lista_paises = df_antes_2000.groupby("country")[["title"]].count()

lista_paises.sort_values(by = "title", ascending=False)
```

#### Ejercicio 13: ¿Cuál es la puntuación promedio en IMDb de los títulos lanzados en 2021 agrupados por país?

```python
df_2021 = df[df["release_year"] == 2021]
df_2021.groupby("country")[["IMDB Score"]].mean().sort_values(by = "IMDB Score", ascending=False)
```

#### Ejercicio 14: ¿Cuántos títulos fueron dirigidos por cada director? Ordena los resultados de mayor a menor

```python
df_titulos_director = df.groupby("director")[["title"]].count()

df_titulos_director.sort_values(by = "title", ascending=False)
```

#### Ejercicio 15: ¿Cuál es el lenguaje más frecuente en los títulos lanzados en los últimos 5 años agrupado por país?

```python
df_ultimos_5_años = df[df['release_year']>=2019]

# Agrupo país, idioma y cuento cantidad 
conteo_idioma = df_ultimos_5_años.groupby(["country", "Language"]).size().reset_index(name="cantidad")
conteo_idioma.sort_values(by = "country").head(10)

#otro groupby por país y coger máx titles
conteo_idioma2 = conteo_idioma.groupby("country")["cantidad"].idxmax()
conteo_idioma2 #índices de máximos

conteo_idioma.iloc[conteo_idioma2].sort_values(by = "cantidad", ascending=False)

```
