```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None) 

df = pd.read_csv("netflix_merge.csv")

df.head(5)
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
lanzamientos2021 = df["release_year"]  == 2021
tipo_movie = df['release_year'] == 2021

filtro_fecha_movie = (lanzamientos2021 & tipo_movie)
df[filtro_fecha_movie]['title'].head()
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
patron_condicion_duracion90min = "90 min"
condicion_duracion90min = df["duration"].str.contains(patron_condicion_duracion90min, case=False, regex = True, na = False)

df_duracion90min = df[condicion_duracion90min]
df_duracion90min[["title","duration"]]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
patron_condicion_india = "India"
condicion_india = df["country"].str.contains(patron_condicion_india, case=False, regex = True, na = False)

condicion_año2018 = df["release_year"] > 2018

df_india_2018 = df[condicion_india & condicion_año2018].head()
df_india_2018[["title","country","release_year"]]
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
lista_clasificación = ["TV-14", "R"]
df_clasificación = df[df["rating"].isin(lista_clasificación)]
df_clasificación[["title", "rating"]]
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
patron_scorsese = "Martin Scorsese"
filtro_scorsese = df["director"].str.contains(patron_scorsese, case=False, regex = True, na = False) 

df_scorsese = df[filtro_scorsese]
print("Scorsese")
print(len(df_scorsese["director"]))

patron_nolan = "Christopher Nolan"
filtro_nolan = df["director"].str.contains(patron_nolan, case=False, regex = True, na = False) 

df_nolan = df[filtro_nolan]
print("Nolan")
print(len(df_nolan["director"]))
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
lista_paises = ["United States", "United Kingdom", "Canada"]
df_paises = df[df["country"].isin(lista_paises)]

df_values = df_paises["Language"].value_counts().reset_index()
df_values

condicion_idiomasconbarra = df_values["Language"].str.contains("/", case=False, regex = True, na = False)
df_values[condicion_idiomasconbarra].sum()

```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
condicion_años = df["release_year"].between(2015,2020, inclusive = "both")
df_años = df[condicion_años]

df_años[["title","release_year"]]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
condicion_imdb = df["IMDB Score"].between(6.5,8.0, inclusive = "both")
df_imdb = df[condicion_años]

df_imdb[["title","IMDB Score"]]
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
lista_duraciones = ["90 min", "150 min"]
df_paises = df[df["duration"].isin(lista_duraciones)]

df_paises[["title","duration"]]
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
patron_regex = "Love"
filtro_love = df["title"].str.contains(patron_regex, case=False, regex = True, na = False)
filtro_tvshow = df["type"] == "TV Show"

df_love_tvshow = df[filtro_love & filtro_tvshow]
df_love_tvshow[["title","type"]]
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
patron_regex = "Johnson"
filtro_johnson = df["director"].str.contains(patron_regex, case=False, regex = True, na = False) 
df_johnson = df[filtro_johnson]
df_johnson[["title","director"]]
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
patron_regex1 = "Documentaries"
condicion_doc = df["listed_in"].str.contains(patron_regex1, case=False, regex = True, na = False) 

patron_regex2 = "Mike Tyson"
condicion_tyson = df["cast"].str.contains(patron_regex2, case=False, regex = True, na = False) 

df_doc_tyson = df[condicion_doc & condicion_tyson]
df_doc_tyson[["title","listed_in","cast"]]
```

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

```python
df_query = df.query("country == 'France' and Runtime > 80")[["title", "country", "Runtime"]]
df_query
```

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

```python
df_query = df.query("type == 'Movie' and release_year < 2010")[["title", "type", "release_year"]]
df_query
```
